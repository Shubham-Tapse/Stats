Q1.) Generate a list of 100 integers containing values between 90 to 130 and store it in the variable `int_list`.
After generating the list, find the following:

  (i) Write a Python function to calculate the mean of a given list of numbers.
      Create a function to find the median of a list of numbers.

  

  (ii) Develop a program to compute the mode of a list of integers.

  

  (iii) Implement a function to calculate the weighted mean of a list of values and their corresponding weights.

  

  (iv) Write a Python function to find the geometric mean of a list of positive numbers.

  

  (v) Create a program to calculate the harmonic mean of a list of values.

  

  (vi) Build a function to determine the midrange of a list of numbers (average of the minimum and maximum).

  

  (vii) Implement a Python program to find the trimmed mean of a list, excluding a certain percentage of
        outliers.

````python
import random
import statistics
import math

# Generating a list of 100 integers between 90 and 130
int_list = [random.randint(90, 130) for _ in range(100)]

# Function to calculate the mean of a list
def calculate_mean(numbers):
    return sum(numbers) / len(numbers)

# Function to find the median of a list
def calculate_median(numbers):
    sorted_numbers = sorted(numbers)
    n = len(sorted_numbers)
    midpoint = n // 2
    if n % 2 == 1:
        return sorted_numbers[midpoint]
    else:
        return (sorted_numbers[midpoint - 1] + sorted_numbers[midpoint]) / 2

# Function to compute the mode of a list
def calculate_mode(numbers):
    return statistics.mode(numbers)

# Function to calculate the weighted mean
def weighted_mean(values, weights):
    return sum(value * weight for value, weight in zip(values, weights)) / sum(weights)

# Function to find the geometric mean of a list of positive numbers
def geometric_mean(numbers):
    return math.exp(sum(math.log(x) for x in numbers) / len(numbers))

# Function to calculate the harmonic mean
def harmonic_mean(numbers):
    return len(numbers) / sum(1 / x for x in numbers)

# Function to determine the midrange
def midrange(numbers):
    return (max(numbers) + min(numbers)) / 2

# Function to calculate the trimmed mean
def trimmed_mean(numbers, percentage):
    n = len(numbers)
    k = int(n * (percentage / 100))
    trimmed_list = sorted(numbers)[k:n-k]
    return sum(trimmed_list) / len(trimmed_list)

# Example usage
print("Mean:", calculate_mean(int_list))
print("Median:", calculate_median(int_list))
print("Mode:", calculate_mode(int_list))
print("Weighted Mean:", weighted_mean(int_list, [1] * 100))  # Example with equal weights
print("Geometric Mean:", geometric_mean([x for x in int_list if x > 0]))  # Ensure no non-positive numbers
print("Harmonic Mean:", harmonic_mean(int_list))
print("Midrange:", midrange(int_list))
print("Trimmed Mean:", trimmed_mean(int_list, 10))  # Example with 10% trimming
````

````python
import random
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Generate the list of 500 integers between 200 and 300
int_list2 = [random.randint(200, 300) for _ in range(500)]

# (i) Visualization
def plot_distributions(data):
    plt.figure(figsize=(15, 5))

    # 1. Frequency & Gaussian distribution
    plt.subplot(1, 3, 1)
    sns.histplot(data, bins=20, kde=True, stat='density', color='blue', edgecolor='black')
    plt.title('Frequency & Gaussian Distribution')
    plt.xlabel('Value')
    plt.ylabel('Density')

    # 2. Frequency smoothened KDE plot
    plt.subplot(1, 3, 2)
    sns.kdeplot(data, color='orange', fill=True, bw_adjust=0.5)
    plt.title('Smoothened KDE Plot')
    plt.xlabel('Value')
    plt.ylabel('Density')

    # 3. Gaussian distribution & smoothened KDE plot
    plt.subplot(1, 3, 3)
    sns.histplot(data, bins=20, kde=True, stat='density', color='blue', edgecolor='black')
    sns.kdeplot(data, color='orange', fill=True, bw_adjust=0.5)
    plt.title('Gaussian Distribution & Smoothened KDE Plot')
    plt.xlabel('Value')
    plt.ylabel('Density')

    plt.tight_layout()
    plt.show()

# Call the plot function
plot_distributions(int_list2)

# (ii) Function to calculate range
def calculate_range(numbers):
    return max(numbers) - min(numbers)

# Calculate range for int_list2
range_result = calculate_range(int_list2)

# (iii) Program to find variance and standard deviation
def calculate_variance_std(numbers):
    mean = np.mean(numbers)
    variance = np.var(numbers, ddof=0)  # Population variance
    std_dev = np.std(numbers, ddof=0)   # Population standard deviation
    return variance, std_dev

variance_result, std_dev_result = calculate_variance_std(int_list2)

# (iv) Function to compute interquartile range (IQR)
def calculate_iqr(numbers):
    q75, q25 = np.percentile(numbers, [75, 25])
    return q75 - q25

# Calculate IQR for int_list2
iqr_result = calculate_iqr(int_list2)

# (v) Program to calculate coefficient of variation
def calculate_coefficient_of_variation(numbers):
    mean = np.mean(numbers)
    std_dev = np.std(numbers, ddof=0)
    return (std_dev / mean) * 100  # expressed as a percentage

cv_result = calculate_coefficient_of_variation(int_list2)

# (vi) Function to find mean absolute deviation (MAD)
def calculate_mad(numbers):
    mean = np.mean(numbers)
    return np.mean(np.abs(numbers - mean))

# Calculate MAD for int_list2
mad_result = calculate_mad(int_list2)

# (vii) Program to calculate quartile deviation
def calculate_quartile_deviation(numbers):
    q75, q25 = np.percentile(numbers, [75, 25])
    return (q75 - q25) / 2  # Quartile deviation is half of IQR

quartile_deviation_result = calculate_quartile_deviation(int_list2)

# (viii) Function to find range-based coefficient of dispersion
def calculate_range_based_cd(numbers):
    return (np.std(numbers, ddof=0) / (max(numbers) - min(numbers))) * 100  # as a percentage

range_based_cd_result = calculate_range_based_cd(int_list2)

# Collecting all results
results_dict2 = {
    "range": range_result,
    
    "variance": variance_result,
    
    "standard_deviation": std_dev_result,
    
    "IQR": iqr_result,
    
    "coefficient_of_variation": cv_result,
    
    "mean_absolute_deviation": mad_result,
    
    "quartile_deviation": quartile_deviation_result,
    
    "range_based_coefficient_of_dispersion": range_based_cd_result
}

# Print the results
print(results_dict2)
````

Q3) Write a Python class representing a discrete random variable with methods to calculate its expected
value and variance.

````python
class DiscreteRandomVariable:
    def __init__(self, values, probabilities):
        if len(values) != len(probabilities):
            raise ValueError("The number of values and probabilities must be the same.")
        if not abs(sum(probabilities) - 1) < 1e-6:
            raise ValueError("The sum of probabilities must be 1.")
        
        self.values = values
        self.probabilities = probabilities

    def expected_value(self):
        return sum(value * prob for value, prob in zip(self.values, self.probabilities))

    def variance(self):
        expected_value = self.expected_value()
        return sum(((value - expected_value) ** 2) * prob for value, prob in zip(self.values, self.probabilities))

# Example usage:
values = [1, 2, 3, 4]
probabilities = [0.1, 0.3, 0.4, 0.2]
rv = DiscreteRandomVariable(values, probabilities)
print("Expected Value:", rv.expected_value())
print("Variance:", rv.variance())
````

Q4) Implement a program to simulate the rolling of a fair six-sided die and calculate the expected value and
variance of the outcomes.


````python
import random

class DieSimulation:
    def __init__(self, sides=6, rolls=10000):
        self.sides = sides
        self.rolls = rolls
        self.outcomes = []

    def roll_die(self):
        # Simulate rolling the die 'rolls' number of times
        self.outcomes = [random.randint(1, self.sides) for _ in range(self.rolls)]
        return self.outcomes

    def calculate_expected_value(self):
        # Calculate expected value for a fair die: (1+2+3+...+6)/6 = 3.5
        return sum(self.outcomes) / len(self.outcomes)

    def calculate_variance(self):
        # Calculate variance based on the outcomes
        expected_value = self.calculate_expected_value()
        return sum((outcome - expected_value) ** 2 for outcome in self.outcomes) / len(self.outcomes)

# Run the simulation
die_simulation = DieSimulation()
die_simulation.roll_die()

# Calculate expected value and variance
expected_value = die_simulation.calculate_expected_value()
variance = die_simulation.calculate_variance()

print("Expected Value:", expected_value)
print("Variance:", variance)
````

Q5) Create a Python function to generate random samples from a given probability distribution (e.g.,
binomial, Poisson) and calculate their mean and variance.

````python
import numpy as np

def generate_samples(distribution, params, sample_size=1000):
    """
    Generates random samples from a specified probability distribution and calculates their mean and variance.
    
    Parameters:
        distribution (str): Type of distribution ('binomial' or 'poisson')
        params (dict): Parameters for the distribution
                       - For 'binomial': {'n': number of trials, 'p': probability of success}
                       - For 'poisson': {'lambda': average rate of events}
        sample_size (int): Number of random samples to generate (default is 1000)
        
    Returns:
        tuple: (mean, variance) of the generated samples
    """
    if distribution == 'binomial':
        n = params.get('n', 10)  # Default number of trials is 10 if not specified
        p = params.get('p', 0.5)  # Default probability of success is 0.5 if not specified
        samples = np.random.binomial(n, p, sample_size)
    
    elif distribution == 'poisson':
        lam = params.get('lambda', 5)  # Default lambda is 5 if not specified
        samples = np.random.poisson(lam, sample_size)
    
    else:
        raise ValueError("Unsupported distribution. Choose 'binomial' or 'poisson'.")
    
    mean = np.mean(samples)
    variance = np.var(samples)
    return mean, variance

# Example usage
# Binomial distribution with n=10 trials and p=0.5 probability of success
binomial_mean, binomial_variance = generate_samples('binomial', {'n': 10, 'p': 0.5})
print("Binomial Distribution - Mean:", binomial_mean, "Variance:", binomial_variance)

# Poisson distribution with lambda=5
poisson_mean, poisson_variance = generate_samples('poisson', {'lambda': 5})
print("Poisson Distribution - Mean:", poisson_mean, "Variance:", poisson_variance)
````

Q6) Write a Python script to generate random numbers from a Gaussian (normal) distribution and compute
the mean, variance, and standard deviation of the samples.

````python
import numpy as np

def generate_gaussian_samples(mean=0, std_dev=1, sample_size=1000):
    """
    Generates random samples from a Gaussian (normal) distribution and computes
    the mean, variance, and standard deviation of the samples.
    
    Parameters:
        mean (float): The mean of the Gaussian distribution (default is 0)
        std_dev (float): The standard deviation of the Gaussian distribution (default is 1)
        sample_size (int): Number of random samples to generate (default is 1000)
        
    Returns:
        tuple: (sample_mean, sample_variance, sample_std_dev) of the generated samples
    """
    # Generate Gaussian samples
    samples = np.random.normal(loc=mean, scale=std_dev, size=sample_size)
    
    # Calculate statistics
    sample_mean = np.mean(samples)
    sample_variance = np.var(samples)
    sample_std_dev = np.std(samples)
    
    return sample_mean, sample_variance, sample_std_dev

# Example usage
mean = 5  # Set the desired mean
std_dev = 2  # Set the desired standard deviation

sample_mean, sample_variance, sample_std_dev = generate_gaussian_samples(mean, std_dev)
print("Sample Mean:", sample_mean)
print("Sample Variance:", sample_variance)
print("Sample Standard Deviation:", sample_std_dev)
````

and 'tip':

  

  (i) Write a Python function that calculates their skewness.

  (iv) Implement a Python program that calculates the Pearson correlation coefficient between two columns.


  (v) Write a script to visualize the correlation between two specific columns in a Pandas DataFrame using
scatter plots.

````python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# Load the 'tips' dataset
tips = sns.load_dataset('tips')

# Select the 'total_bill' and 'tip' columns
total_bill = tips['total_bill']
tip = tips['tip']

# (i) Function to Calculate Skewness
def calculate_skewness(data):
    return skew(data)

# Calculate skewness for 'total_bill' and 'tip'
skew_total_bill = calculate_skewness(total_bill)
skew_tip = calculate_skewness(tip)
print("Skewness of 'total_bill':", skew_total_bill)
print("Skewness of 'tip':", skew_tip)

# (ii) Program to Determine Skewness Type
def determine_skewness_type(skew_value):
    if skew_value > 0.5:
        return "Positive skewness"
    elif skew_value < -0.5:
        return "Negative skewness"
    else:
        return "Approximately symmetric"

# Determine skewness type for 'total_bill' and 'tip'
skewness_type_total_bill = determine_skewness_type(skew_total_bill)
skewness_type_tip = determine_skewness_type(skew_tip)
print("Skewness type of 'total_bill':", skewness_type_total_bill)
print("Skewness type of 'tip':", skewness_type_tip)

# (iii) Function to Calculate Covariance
def calculate_covariance(col1, col2):
    return np.cov(col1, col2, ddof=0)[0][1]

# Calculate covariance between 'total_bill' and 'tip'
covariance_total_bill_tip = calculate_covariance(total_bill, tip)
print("Covariance between 'total_bill' and 'tip':", covariance_total_bill_tip)

# (iv) Program to Calculate Pearson Correlation Coefficient
def calculate_pearson_correlation(col1, col2):
    return np.corrcoef(col1, col2)[0][1]

# Calculate Pearson correlation between 'total_bill' and 'tip'
pearson_corr_total_bill_tip = calculate_pearson_correlation(total_bill, tip)
print("Pearson correlation coefficient between 'total_bill' and 'tip':", pearson_corr_total_bill_tip)

# (v) Script to Visualize Correlation with Scatter Plot
def plot_scatter(df, col1, col2):
    plt.figure(figsize=(8, 6))
    plt.scatter(df[col1], df[col2], alpha=0.6, color='blue')
    plt.title(f'Scatter Plot between {col1} and {col2}')
    plt.xlabel(col1)
    plt.ylabel(col2)
    plt.show()

# Plot scatter between 'total_bill' and 'tip'
plot_scatter(tips, 'total_bill', 'tip')
````

Q8) Write a Python function to calculate the probability density function (PDF) of a continuous random
variable for a given normal distribution.


````python
import numpy as np
from scipy.stats import norm

def normal_pdf(x, mean=0, std_dev=1):
    """
    Calculate the PDF of a normal distribution at a given point x.
    
    Parameters:
        x (float): The point at which to evaluate the PDF.
        mean (float): The mean (mu) of the normal distribution (default is 0).
        std_dev (float): The standard deviation (sigma) of the normal distribution (default is 1).
    
    Returns:
        float: The probability density at point x.
    """
    # Method 1: Using the formula for normal distribution PDF
    pdf_value_formula = (1 / (std_dev * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - mean) / std_dev) ** 2)
    
    # Method 2: Using scipy's norm.pdf
    pdf_value_scipy = norm.pdf(x, loc=mean, scale=std_dev)
    
    return pdf_value_formula, pdf_value_scipy

# Example usage
x = 1
mean = 0
std_dev = 1

pdf_formula, pdf_scipy = normal_pdf(x, mean, std_dev)
print("PDF using formula:", pdf_formula)
print("PDF using scipy:", pdf_scipy)
````

Q9) Create a program to calculate the cumulative distribution function (CDF) of exponential distribution.


````python
import numpy as np
from scipy.stats import expon

def exponential_cdf(x, rate=1):
    """
    Calculate the CDF of an exponential distribution at a given point x.
    
    Parameters:
        x (float): The point at which to evaluate the CDF.
        rate (float): The rate parameter (lambda) of the exponential distribution (default is 1).
    
    Returns:
        tuple: The CDF value at point x using the formula and scipy method.
    """
    # Method 1: Using the formula for exponential distribution CDF
    if x < 0:
        cdf_value_formula = 0
    else:
        cdf_value_formula = 1 - np.exp(-rate * x)
    
    # Method 2: Using scipy's expon.cdf
    cdf_value_scipy = expon.cdf(x, scale=1/rate)  # scale is the inverse of rate (1/lambda)
    
    return cdf_value_formula, cdf_value_scipy

# Example usage
x = 1
rate = 0.5

cdf_formula, cdf_scipy = exponential_cdf(x, rate)
print("CDF using formula:", cdf_formula)
print("CDF using scipy:", cdf_scipy)
````

Q10) Write a Python function to calculate the probability mass function (PMF) of Poisson distribution.

````python
import numpy as np
from scipy.stats import poisson
from math import exp, factorial

def poisson_pmf(k, rate):
    """
    Calculate the PMF of a Poisson distribution at a given point k.
    
    Parameters:
        k (int): The number of occurrences (non-negative integer).
        rate (float): The average rate of occurrences (lambda).
    
    Returns:
        tuple: PMF value at point k using the formula and scipy method.
    """
    # Method 1: Using the formula for Poisson distribution PMF
    if k < 0:
        raise ValueError("k must be a non-negative integer")
    pmf_value_formula = (rate ** k * exp(-rate)) / factorial(k)
    
    # Method 2: Using scipy's poisson.pmf
    pmf_value_scipy = poisson.pmf(k, rate)
    
    return pmf_value_formula, pmf_value_scipy

# Example usage
k = 3  # Number of occurrences
rate = 2.5  # Average rate (lambda)

pmf_formula, pmf_scipy = poisson_pmf(k, rate)
print("PMF using formula:", pmf_formula)
print("PMF using scipy:", pmf_scipy)
````

Q11) A company wants to test if a new website layout leads to a higher conversion rate (percentage of visitors
who make a purchase). They collect data from the old and new layouts to compare.


To generate the data use the following command:

```python

import numpy as np

# 50 purchases out of 1000 visitors

old_layout = np.array([1] * 50 + [0] * 950)

# 70 purchases out of 1000 visitors  

new_layout = np.array([1] * 70 + [0] * 930)

  ```

Apply z-test to find which layout is successful.

````python
import numpy as np
from statsmodels.stats.proportion import proportions_ztest

# Generate data
old_layout = np.array([1] * 50 + [0] * 950)
new_layout = np.array([1] * 70 + [0] * 930)

# Calculate conversion numbers
successes = [old_layout.sum(), new_layout.sum()]  # Number of purchases
n_obs = [len(old_layout), len(new_layout)]  # Number of visitors

# Perform Z-test
z_stat, p_value = proportions_ztest(successes, n_obs, alternative='larger')

print("Z-statistic:", z_stat)
print("P-value:", p_value)

# Interpret the result
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new layout has a higher conversion rate.")
else:
    print("Fail to reject the null hypothesis: No significant difference in conversion rates.")
````

Q14) A customer service department claims that their average response time is less than 5 minutes. A sample
of recent customer interactions was taken, and the response times were recorded.


Implement the below code to generate the array of response time:

```python

response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

```

Implement z-test to find the claims made by customer service department are tru or false.

````python
import numpy as np
import scipy.stats as stats

# Generate sample data
response_times = np.array([4.3, 3.8, 5.1, 4.9, 4.7, 4.2, 5.2, 4.5, 4.6, 4.4])

# Define the population mean under the null hypothesis
mu_0 = 5.0

# Calculate sample mean and standard deviation
sample_mean = np.mean(response_times)
sample_std = np.std(response_times, ddof=1)  # Use ddof=1 for sample standard deviation
n = len(response_times)

# Calculate the Z-score
z_score = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# Calculate the p-value for the one-tailed test
p_value = stats.norm.cdf(z_score)

# Output results
print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)
print("Z-score:", z_score)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The average response time is significantly less than 5 minutes.")
else:
    print("Fail to reject the null hypothesis: No significant evidence that the average response time is less than 5 minutes.")
````

Q15) A company is testing two different website layouts to see which one leads to higher click-through rates.
Write a Python function to perform an A/B test analysis, including calculating the t-statistic, degrees of
freedom, and p-value.


Use the following data:

```python

layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]

layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

````python
import numpy as np
from scipy import stats

def ab_test(layout_a, layout_b):
    # Calculate sample means
    mean_a = np.mean(layout_a)
    mean_b = np.mean(layout_b)
    
    # Calculate sample standard deviations
    std_a = np.std(layout_a, ddof=1)  # Sample standard deviation
    std_b = np.std(layout_b, ddof=1)  # Sample standard deviation
    
    # Calculate sample sizes
    n_a = len(layout_a)
    n_b = len(layout_b)
    
    # Calculate the t-statistic
    t_statistic = (mean_a - mean_b) / np.sqrt((std_a**2 / n_a) + (std_b**2 / n_b))
    
    # Calculate degrees of freedom
    df = n_a + n_b - 2
    
    # Calculate the p-value for the two-tailed test
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    return t_statistic, df, p_value

# Given data
layout_a_clicks = [28, 32, 33, 29, 31, 34, 30, 35, 36, 37]
layout_b_clicks = [40, 41, 38, 42, 39, 44, 43, 41, 45, 47]

# Perform A/B test
t_statistic, df, p_value = ab_test(layout_a_clicks, layout_b_clicks)

# Output results
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", df)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant difference in click-through rates between the two layouts.")
else:
    print("Fail to reject the null hypothesis: No significant difference in click-through rates.")
````

new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

'''

````python
import numpy as np
from scipy import stats

def compare_drugs(existing_drug, new_drug):
    # Calculate sample means
    mean_existing = np.mean(existing_drug)
    mean_new = np.mean(new_drug)
    
    # Calculate sample standard deviations
    std_existing = np.std(existing_drug, ddof=1)  # Sample standard deviation
    std_new = np.std(new_drug, ddof=1)  # Sample standard deviation
    
    # Calculate sample sizes
    n_existing = len(existing_drug)
    n_new = len(new_drug)
    
    # Calculate the t-statistic
    t_statistic = (mean_existing - mean_new) / np.sqrt((std_existing**2 / n_existing) + (std_new**2 / n_new))
    
    # Calculate degrees of freedom
    df = n_existing + n_new - 2
    
    # Calculate the p-value for the one-tailed test
    p_value = stats.t.cdf(t_statistic, df=df)
    
    return t_statistic, df, p_value

# Given cholesterol levels
existing_drug_levels = [180, 182, 175, 185, 178, 176, 172, 184, 179, 183]
new_drug_levels = [170, 172, 165, 168, 175, 173, 170, 178, 172, 176]

# Perform the comparison
t_statistic, df, p_value = compare_drugs(existing_drug_levels, new_drug_levels)

# Output results
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", df)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The new drug is significantly more effective than the existing drug.")
else:
    print("Fail to reject the null hypothesis: No significant difference in effectiveness between the drugs.")
````

Q17) A school district introduces an educational intervention program to improve math scores. Write a Python
function to analyze pre- and post-intervention test scores, calculating the t-statistic and p-value to
determine if the intervention had a significant impact.


Use the following data of test score:


  ```python

  pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]

  post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
  '''

````python
import numpy as np
from scipy import stats

def analyze_intervention(pre_scores, post_scores):
    # Calculate the differences between post and pre scores
    differences = np.array(post_scores) - np.array(pre_scores)
    
    # Calculate mean and standard deviation of the differences
    mean_difference = np.mean(differences)
    std_difference = np.std(differences, ddof=1)  # Sample standard deviation
    n = len(differences)  # Number of paired observations
    
    # Calculate the t-statistic
    t_statistic = mean_difference / (std_difference / np.sqrt(n))
    
    # Calculate degrees of freedom
    df = n - 1
    
    # Calculate the p-value for the one-tailed test
    p_value = 1 - stats.t.cdf(t_statistic, df=df)
    
    return t_statistic, df, p_value

# Given test scores
pre_intervention_scores = [80, 85, 90, 75, 88, 82, 92, 78, 85, 87]
post_intervention_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

# Perform the analysis
t_statistic, df, p_value = analyze_intervention(pre_intervention_scores, post_intervention_scores)

# Output results
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", df)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: The intervention had a significant positive impact on math scores.")
else:
    print("Fail to reject the null hypothesis: No significant impact of the intervention on math scores.")
````

Q18) An HR department wants to investigate if there's a gender-based salary gap within the company. Develop
a program to analyze salary data, calculate the t-statistic, and determine if there's a statistically
significant difference between the average salaries of male and female employees.


Use the below code to generate synthetic data:


```python

# Generate synthetic salary data for male and female employees

np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)

female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

````python
import numpy as np
from scipy import stats

# Generate synthetic salary data for male and female employees
np.random.seed(0)  # For reproducibility

male_salaries = np.random.normal(loc=50000, scale=10000, size=20)
female_salaries = np.random.normal(loc=55000, scale=9000, size=20)

def analyze_salary_gap(male_salaries, female_salaries):
    # Calculate sample means
    mean_male = np.mean(male_salaries)
    mean_female = np.mean(female_salaries)
    
    # Calculate sample standard deviations
    std_male = np.std(male_salaries, ddof=1)  # Sample standard deviation
    std_female = np.std(female_salaries, ddof=1)  # Sample standard deviation
    
    # Calculate sample sizes
    n_male = len(male_salaries)
    n_female = len(female_salaries)
    
    # Calculate the t-statistic
    t_statistic = (mean_male - mean_female) / np.sqrt((std_male**2 / n_male) + (std_female**2 / n_female))
    
    # Calculate degrees of freedom
    df = n_male + n_female - 2
    
    # Calculate the p-value for the two-tailed test
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df=df))
    
    return t_statistic, df, p_value

# Perform the analysis
t_statistic, df, p_value = analyze_salary_gap(male_salaries, female_salaries)

# Output results
print("T-statistic:", t_statistic)
print("Degrees of Freedom:", df)
print("P-value:", p_value)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a statistically significant difference in average salaries between male and female employees.")
else:
    print("Fail to reject the null hypothesis: No statistically significant difference in average salaries.")
````

Q21) A political analyst wants to determine if there is a significant association between age groups and voter
preferences (Candidate A or Candidate B). They collect data from a sample of 500 voters and classify
them into different age groups and candidate preferences. Perform a Chi-Square test to determine if
there is a significant association between age groups and voter preferences.


Use the below code to generate data:

```python

np.random.seed(0)

age_groups = np.random.choice([ 18 30 , 31 50 , 51+', 51+'], size=30)

voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=30)

 '''

````python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency

# Generate data
np.random.seed(0)

# Create random age groups
age_groups = np.random.choice(['18-30', '31-50', '51+'], size=500)

# Create random voter preferences
voter_preferences = np.random.choice(['Candidate A', 'Candidate B'], size=500)

# Create a DataFrame
data = pd.DataFrame({'Age Group': age_groups, 'Voter Preference': voter_preferences})

# Create a contingency table
contingency_table = pd.crosstab(data['Age Group'], data['Voter Preference'])

# Perform the Chi-Square test
chi2_stat, p_value, dof, expected = chi2_contingency(contingency_table)

# Output results
print("Chi-Square Statistic:", chi2_stat)
print("P-value:", p_value)
print("Degrees of Freedom:", dof)
print("Expected Frequencies:\n", expected)

# Interpret the results
alpha = 0.05  # Significance level
if p_value < alpha:
    print("Reject the null hypothesis: There is a significant association between age groups and voter preferences.")
else:
    print("Fail to reject the null hypothesis: No significant association between age groups and voter preferences.")
````

version from a random sample of customers. Perform an ANOVA test to determine if there is a significant
difference in customer satisfaction scores.


  Use the following data:

  ```python

  # Sample data: Customer satisfaction scores for each product version

  standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]

  premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]

  deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
